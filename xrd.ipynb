# Diffraction des rayons X du MgCl2 (mp-570259)

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plt
```

```python
with MPRester("qpcDzlSRWPfs7J19") as m:
    struct = m.get_structure_by_material_id("mp-570259")
```

```python
xrdcal = XRDCalculator('CuKa')
patt = xrdcal.get_pattern(struct)
```

```python
print("Premier pic du diffractogramme pour un rayonnement X incident CuKa de longueur d’onde λ=1.54060 Å")
print("-------------------------------------------------------------------------------------------------")
print(" ")
print("Angle 2\u0398 [°] :", patt.x[0])
print("Intensité du pic [%] :", patt.y[0])
print("Indices hkl associés :", patt.hkls[0])
print("                   d :", patt.d_hkls[0])
```

```python
print("Deuxième pic du diffractogramme pour un rayonnement X incident CuKa de longueur d’onde λ=1.54060 Å")
print("--------------------------------------------------------------------------------------------------")
print(" ")
print("Angle 2\u0398 [°] :", patt.x[1])
print("Intensité du pic [%] :", patt.y[1])
print("Indices hkl associés :", patt.hkls[1])
print("                   d :", patt.d_hkls[1])
```

```python
print("Troisième pic du diffractogramme pour un rayonnement X incident CuKa de longueur d’onde λ=1.54060 Å")
print("---------------------------------------------------------------------------------------------------")
print(" ")
print("Angle 2\u0398 [°] :", patt.x[2])
print("Intensité du pic [%] :", patt.y[2])
print("Indices hkl associés :", patt.hkls[2])
print("                   d :", patt.d_hkls[2])
```

```python
print("Trois premiers pics du diffractogramme pour un rayonnement X incident CuKa de longueur d’onde λ=1.54060 Å")
xrdcal.show_plot(struct,two_theta_range=(10,30),annotate_peaks=True,ax=None,with_labels=True)
```
