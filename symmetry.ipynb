```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.symmetry.analyzer import *
from pymatgen.io.cif import *
from pymatgen.core.operations import *
from tabulate import tabulate
from pymatgen.symmetry.site_symmetries import *
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
with MPRester("qpcDzlSRWPfs7J19") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-570259")
```

```python
# Opérations de symétrie
a = SpacegroupAnalyzer(structure).get_symmetry_operations()
Sym_1 = a[2]
Sym_2 = a[5]
Sym_3 = a[7]
#print(Sym_1)
#print(Sym_2)
#print(Sym_3)

# Différents atomes
Mg1_i = [0.0,1.0,1.0]#6
Mg2_i = [0.0,0.0,1.0]#4
Cl1_i = [0.333333,0.666667,0.213751]#2
#print(Mg1_i)
#print(Mg2_i)
#print(Cl1_i)

# Application des opérations de symétrie
Mg1_f = Sym_1.operate(Mg1_i)
Mg2_f = Sym_2.operate(Mg2_i)
Cl1_f = Sym_3.operate(Cl1_i)
#print(Mg1_f)
#print(Mg2_f)
#print(Cl1_f)
```

```python
print("Première opération de symétrie")
print("------------------------------")
print(" ")
print("Atome de départ : Mg", Mg1_i)
print(" ")
print("Opération de symétrie : axe C3 : ")
print(Sym_1)
print(" ")
print("Atome d'arrivée : Mg", Mg1_f)
print(" ")
print("Image :")

image_1=img.imread('Mg1_Op1.png')
fig, axes = plt.subplots(figsize=(10,10))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(image_1)
plt.show()
```

```python
print("Deuxième opération de symétrie")
print("------------------------------")
print(" ")
print("Atome de départ : Mg", Mg2_i)
print(" ")
print("Opération de symétrie : axe 3-bar : ")
print(Sym_2)
print(" ")
print("Atome d'arrivée : Mg", Mg2_f)
print(" ")
print("Image :")

image_2=img.imread('Mg2_Op2.png')
fig, axes = plt.subplots(figsize=(10,10))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(image_2)
plt.show()
```

```python
print("Troisième opération de symétrie")
print("-------------------------------")
print(" ")
print("Atome de départ : Cl", Cl1_i)
print(" ")
print("Opération de symétrie : plan miroir : ")
print(Sym_3)
print(" ")
print("Atome d'arrivée : Cl", Cl1_f)
print(" ")
print("Image :")

image_3=img.imread('Cl1_Op3.png')
fig, axes = plt.subplots(figsize=(10,10))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(image_3)
plt.show()
```
